1.10 Manejo de exepciones

El lidiar con errores de programación tiene (al menos) dos partes:
1. La primera es cuando te metes en problemas porque tu codigo, aparentemente correcto. se alimenta con datos incorrectos. Por ejemplo, esperas que se ingrese el codigo un valor entero, pero el usuario ingresa algunas legras al azar. Otro escenario es cuando intentas abrir un archivo ya no existe o se le cambio el nombre.

2. la segunda parte de lidiar con errores de programacion se revela cuando ocurre un comportamiendo deseado del programa debido a errores que se cometieron cuando se estaba escribiendo el codigo. Este tipo de error se denomina comunmente "bug"

```python
x = float(input("introduce un numero"))
y = 5
print("la multiplicacion de ", y, "por",x, "es igual a", y*x,"\n")

```

```python
while(True):
  try:
      x = float(input("introduce un numero: "))
      y = 5
      print("la multiplicacion de ", y, "por",x, "es igual a", y*x,"\n")
      break
  except:
    print("Ha ocurrido un error, introduce bien el numero")


```

```python
while(True):
  try:
      x = float(input("introduce un numero: "))
      y = 5
      print("la multiplicacion de ", y, "por",x, "es igual a", y*x,"\n")
      break
  except:
    print("Ha ocurrido un error, introduce bien el numero")
  else:
    print("Todo ha funcionado correctamente")
    break
  finally:
    print("Fin del bucle")


```

```python
from logging import exception
while(True):
  try:
      x = float(input("introduce un numero: "))
      y = 100
      print("la divicion de ", y, "por",x, "es igual a", y/x,"\n")
      break
  except ValueError:
    print("Ha ocurrido un error, introduce bien el numero")
  except exception as e:
    print("Error", type(e).__name__)
  except ZeroDivisionError:
    print("No se puede dividir entre cero")
  else:
    print("Todo ha funcionado correctamente")
    break
  finally:
    print("Fin del bucle")


```
